## 信号
### 定义
信号的定义：随时间（空间）变化的物理量。可以表示为一个或多个独立变量的函数。如 x(t) 
### 分类
确定和随机(如噪声)；  
连续和离散（肉眼可判断）；  
周期和非周期：简单的函数肉眼就可判断。可能考叠加形式的判断，如$sin(3\pi t) + cos(2t)——不是周期信号$  
>————会考离散、周期信号的判断。见题1.2

### 信号的能量与功率
对于信号$x(t)$有如下定义：
瞬时功率：$$P(t) = x^2(t) \ or\   |x(t)|^2$$
$[t_1,t_2]$时期内的能量: $$\int_{t_1}^{t_2}p(t)dt$$
总能量$$E_\infty = \int_{-\infty}^{\infty}p(t)dt$$
（总）平均功率$$P_\infty =\lim_{T\rightarrow \infty} \frac{\int_{-T}^{T} P(t)dt } {2T}$$
（这里是先定义的功率，再用功率定义了能量）   
根据能量的情况，信号可以分为：  
能量有限信号(能量信号):$E_\infty<\infty$
功率有限信号(功率信号):$P<\infty$

## 阶跃响应和冲激响应

### 求解h(t)
（题目都是__n阶线性齐次__的系统）  
对于方程$$D(p)h(t) = N(p)\delta(t)$$
即$$(a_np^n + a_{n-1}p^{n-1} + \dots + a_1p + a_0)h(t) = (b_mp^m + b_{m-1}p^{m-1} + \dots + b_1p + b_0)\delta(t)$$
即$$h(t) = \frac{(a_np^n + a_{n-1}p^{n-1} + \dots + a_1p + a_0)}{(b_mp^m + b_{m-1}p^{m-1} + \dots + b_1p + b_0)}\delta(t)$$ 

先假设根都是单根。
若$m>n$则先把$H(p)$化成真分式。
若$n>m$则将$H(p)$进行部分分式展开。  
$$h(t) = [\frac{k_1}{p-\lambda_1} + \frac{k_2}{p - \lambda_2} + \dots + \frac{k_n}{p-\lambda_n}]\delta(t)\\
 = \frac{k_1}{p-\lambda_1}\delta(t) + \frac{k_2}{p - \lambda_2}\delta(t) + \dots + \frac{k_n}{p-\lambda_n}\delta(t)
$$
相当于方程分为了这些线性叠加。  
那么每一个小分量的解就是一个一次特征根的解，如  
$$h_1(t) = \frac{k_1}{p-\lambda_1}\delta(t) 或 \frac{d}{dt}h_1(t) - \lambda_1h(t) = k_1\delta(t)\\这个小分量的解是\\
h_1(t) = k_1e^{\lambda_1t}\epsilon(t)
$$

总响应就是这些分量线性叠加  
$$
h(t) = \sum_{i=1}^{n}k_ie^{\lambda_it}\epsilon(t)。
每个系数k待定$$
即$$r_{zi} = c_1e^{\lambda_1t} + c_2e^{\lambda_2t} + \dots + c_n e^{t_nt} , t\ge 0^-$$

如果$m>n$，则化出来会有多项式分量

## 卷积及其性质
### 几何解释
卷积的表达式：$$f(t)*g(t) = \int_{-\infty}^{\infty}f(\tau-t)g(\tau)d\tau$$
（1）反褶(左右对称)（2）位移（3）相乘：（4）积分    
here's the illustration:
![Convolution Illustration](convolution.png)
### 卷积的性质
- 交换律。结合律。分配律。
- 积分与微分性质(__！重要__)：  
If $r(t) = e(t)*h(t) = e(t)*h(t)$ then $$\frac{d}{dt}r(t) =\frac{d}{dt}e(t) *h(t) = \frac{d}{dt}h(t) *e(t) $$   
- 奇异函数的卷积：  
函数与$\delta(t)$的卷积：$$f(t)*\delta(t)=f(t)$$。——__函数卷$\delta(t)$等于它自己__！  
其他奇异函数的卷积可用微分性质求。
- 函数延时的卷积
$$e(t-t_0)*h(t-t_1) = e(t-t_0-t_1)*h(t) = e(t)*h(t-t_0-t_1)$$
——延时可以随便往人身上丢，反正最后相当于结果统一延时。
(速证：因为时延相当于卷一个$\delta(t-t_0)$,而卷是可以变顺序的，所以随便谁时延，最后加在一起时延。)

## 反变换
——课堂范围内要求掌握有理函数：

$$F(s) = \frac{N(s)}{D(s)} 
= \frac{b_ms^m + b_{m-1}s^{m-1}...}
    {b_ms^n + a_{m-1}s^{m-1}...}
$$的反变换。通法是用留数求。  
下面介绍对部分分式求反变换的方法：

### 部分分式展开法
首先，给到一个部分分式，若是假分式则先化为真分式。使用_短除法_。  
然后按部分分式的类型分多种情况：
$D(s) = 0$的根$s_1, s_2 .....s_n$即为$F(s)$的极点
-  无重根
无重根，则可以表示为$$F(s) = \frac{N(s)}{D(s)} = \frac{N(s)}{(s-s_1)(s-s_2)\dots (s-s_n)}$$
亦$$F(s) = \frac{K_1}{s-s_1} + \frac{K_2}{s-s_2} + \dots + \frac{K_n}{s-s_n}$$
求$K_i$:式子左右乘$(s-s_i)$。。。。。  
结果：$$k_ie^{s_i}\epsilon(t) \leftrightarrow \frac{k_i}{s-s_i}$$
若存在虚根，一种做法是上述做法直接沿用，结果带复数。另一种做法是__配出$cos$和$sin$的反变换式子__:
$$\sin(t)\epsilon(t)\leftrightarrow \frac{\omega_0}{s^2+\omega_o^2}$$  
$$\cos(t)\epsilon(t)\leftrightarrow \frac{s}{s^2+\omega_o^2}$$  
$$举例:F(s) = \frac{s}{s^2 + 2s + 2} = \frac{s+1}{(s+1)^2 + 1} + \frac{-1}{(s+1)^2 + 1}\\
\therefore f(t) = e^{-t}(cos(t) - sin(t))\epsilon(t)（e^{-t}是用时移性质来的）$$

- 有重根
单根项还是照样求。重根项的公式：
$$K_{ik} = \frac{1}{(p-k)!}\frac{d^{p-k}}{ds^{p-k}}[(s-s_k)\frac{N(s)}{D(s)}]$$
$$例如二重根的两个系数：K_{i1} = [(s-s_k)\frac{N(s)}{D(s)}]K_{i2} = \frac{d}{ds}[(s-s_k)\frac{N(s)}{D(s)}]$$
（k表示这个部分分式是第几重）
放个全面的例子：
$$F(s) = \frac{1}{s(s+1)^2(s^+2s+2)}\\ =\frac{k_1}{s} + \frac{k_{21}}{s+1} (重根写开两个)+ \frac{k_{22}}{s+1} + \frac{k_3}{(s-1-j)} + \frac{k_4}{(s-1+j)}$$
（不好描述的话，总之见题就行了。）


### 留数法
对一阶极点：$$Res_k = [(s-s_k)F(s)e^{st}]_{s=s_k}$$
p阶极点:$$Res_k = \frac{1}{(p-1)!}[\frac{d^{p-1}}{ds^{p-1}}(s-s_k)^pF(s)e^{st}]$$
（其实也好背。外面分母是$p-1$的阶乘，所以0和1都没区别；求导也是$p-1$阶，二重根就一阶导；乘$(s-s_k)$是肯定的）
最后，由留数定理反变换的结果等于留数之和。即$$f(t) = \sum Res(s_k)$$
